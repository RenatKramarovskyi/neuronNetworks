2. Нормалізуавати дані за допомогою MinMaxScaler, приклад на
зображенні.


```python


import pandas as pd
import numpy as np
import plotly.graph_objects as go

from sklearn.preprocessing import MinMaxScaler

dataset = {
    'age': np.random.randint(low = 18, high = 100, size=(1,100)).flatten(),
    'health_rate': np.random.randint(low = 1, high = 11, size=(1,100)).flatten(),
    'purchase_price': np.round(np.random.uniform(low = 10, high = 1000, size=(1, 100)), 2).flatten(),
    'monthly_income' : np.random.randint(low = 500, high = 2000, size=(1,100)).flatten()
}

df = pd.DataFrame(dataset)
# print(df)

fig_before = go.Figure()
for column in df.columns:
    fig_before.add_trace(go.Scatter(x = df.index, y = df[column], mode='lines',
                                    name=column.capitalize().replace("_", " ")))

fig_before.update_layout(title='Data before Normalization',
                         xaxis_title='position',
                         yaxis_title='value')
fig_before.show()

#=============================================================
scalerObj = MinMaxScaler()
df = pd.DataFrame(scalerObj.fit_transform(df), columns=df.columns)

fig_normalized = go.Figure()

for column in df.columns:
    fig_normalized.add_trace(go.Scatter(x = df.index, y = df[column], mode='lines',
                                        name=column.capitalize().replace("_", " ")))

fig_normalized.update_layout(title='Data after Normalization',
                             xaxis_title='position',
                             yaxis_title='value')

fig_normalized.show()
```

3. Згенерувати 2 розподіли з різними значеннями проміжку за
допомогою np.random.normal.

```python
import numpy as np
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler

votes1 = np.clip(np.random.normal(loc = 50, scale = 20, size = 800), 0, 100)
votes2 = np.clip(np.random.normal(loc = 100, scale = 50, size = 800), 0, 200)


fig1 = go.Figure()
fig1.add_trace(go.Histogram(x = votes1, name = '1st votes data'))
fig1.update_layout(title = '1st votes data', xaxis_title = 'votes', yaxis_title = 'frequency')
fig1.show()


fig2 = go.Figure()
fig2.add_trace(go.Histogram(x = votes2, name = '2nd votes data'))
fig2.update_layout(title = '2nd votes data', xaxis_title = 'votes', yaxis_title = 'frequency')
fig2.show()

```

4. Стандартизувати попередньо згенеровані розподіли за
допомогою StandardScaler та відобразити на графіку

```python
from sklearn.preprocessing import StandardScaler

scalerObj = StandardScaler()

standardized_votes1 = scalerObj.fit_transform(votes1.reshape(-1, 1)).flatten()
standardized_votes2 = scalerObj.fit_transform(votes2.reshape(-1, 1)).flatten()
# ===================
fig1 = go.Figure()
fig1.add_trace(go.Histogram(x = votes1, name = 'Original votes data 1'))

fig1.update_layout(
    title  ='Original votes data 1',
    xaxis_title = 'votes', yaxis_title = 'dens'
)

fig1.add_trace(go.Histogram(x = standardized_votes1, name = 'Standardized votes data 1'))

fig1.update_layout(
    barmode = 'overlay',
    title_text = 'Data of votes 1 before and after Standardization'
)
# ===================
fig2 = go.Figure()
fig2.add_trace(go.Histogram(x = votes2, name = 'Original votes data 2'))

fig2.update_layout(
    title  ='Original votes data 2',
    xaxis_title = 'votes', yaxis_title = 'dens'
)

fig2.add_trace(go.Histogram(x = standardized_votes2, name = 'Standardized votes data 2'))

fig2.update_layout(
    barmode = 'overlay',
    title_text = 'Data of votes 2 before and after Standardization'
)
# ===================
fig1.show()
fig2.show()
```

```python
import numpy as np
import pandas as pd
import plotly.express as pe
import tensorflow as tsf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split



df = pd.read_csv('sample_data/alarms_no_scalse.csv')
print(df)
# canv = pe.scatter_3d(df, x='sound', y='distance', z='alarm')
canv = pe.scatter_3d(df, x='sound', y='visibility', z='alarm', color='alarm', color_continuous_scale='jet')
canv.show()
# obviously alarm more then, when more soud and less distance
# or alarm more then, twhen soun and visibility more

X = df[['sound', 'distance', 'visibility']].values
Y = df['alarm'].values
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3) #70\30 train\test

model = models.Sequential([
    layers.Dense(1, input_shape=(X_train.shape[1],), activation='sigmoid'),
    # layers.Dense(1, activation='sigmoid')
    # layers.Dense(8, input_shape=(X_train.shape[1],), activation='relu')
    # layers.Dense(1, activation='relu')
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])

model.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_test, Y_test))

loss, accuracy = model.evaluate(X_test, Y_test)
print(f"""
Total test lost : {loss}
Total test accuracy : {accuracy}
""")
```

```python
import numpy as np
import pandas as pd
import plotly.express as pe
import tensorflow as tsf
from tensorflow.keras import layers, models
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import MinMaxScaler
import joblib

print("Before normalization : ")
print(df)

scale_list = ['distance', 'sound', 'visibility']
scalerObj = MinMaxScaler()
df[scale_list] = scalerObj.fit_transform(df[scale_list])

joblib.dump(scalerObj, 'min_max_scale.pkl')
scaled_df = pd.DataFrame(df, columns = df.columns)
print("After normalization : ")
print(scaled_df)

X = scaled_df[['sound', 'distance', 'visibility']].values
Y = scaled_df['alarm'].values

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3) #70\30 train\test
model = models.Sequential([
    layers.Dense(1, input_shape=(X_train.shape[1],), activation='sigmoid'),
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mse'])
model.fit(X_train, Y_train, epochs=50, batch_size=32, validation_data=(X_test, Y_test))

loss, accuracy = model.evaluate(X_test, Y_test)
print(f"""
Total test lost : {loss}
Total test accuracy : {accuracy}
""")
```
